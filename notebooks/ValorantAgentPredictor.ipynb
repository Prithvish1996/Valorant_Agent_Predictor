````python
import pandas as pd
import numpy as np
````

---

### Phase 2: Data Loading

**Objective:** Load source datasets into memory for processing.

**Justification:** Loading both datasets separately allows inspection before integration. This ensures data integrity and enables early detection of format issues.


````python
games_df = pd.read_csv('../Dataset/valorant_games.csv')
agents_df = pd.read_csv('../Dataset/agent_master_list.csv')

print(f"Match History: {len(games_df)} records")
print(f"Agent Reference: {len(agents_df)} agents")
````

---

### Phase 3: Data Integration

**Objective:** Merge match history with agent role information.

**Justification:** Left join preserves all match records while enriching them with role classifications. This creates a unified dataset enabling role-based analysis as required by business objectives.


````python
df = games_df.merge(agents_df, left_on='agent', right_on='Agent', how='left')
df = df.drop(columns=['Agent'])
df = df.rename(columns={'Role': 'role'})

print(f"Integrated Dataset: {len(df)} records")
print(f"Columns: {list(df.columns)}")
````

---

### Phase 4: Feature Engineering - Performance Score

**Objective:** Create a performance metric combining win outcome and KDR.

**Justification:** Raw win/loss is binary and doesn't capture performance quality. The tiered scoring system (0, 0.5, 1.0) rewards strong performances while filtering unreliable data points with low KDR.

| Score | Criteria | Interpretation |
|-------|----------|----------------|
| 1.0 | Win + KDR ≥ 0.7 | Strong performance |
| 0.5 | Win + KDR ≥ 0.4 | Acceptable performance |
| 0.0 | Loss or KDR < 0.4 | Excluded from analysis |


````python
def calculate_performance(row):
    if row['outcome'] == 'Win' and row['kdr'] >= 0.4:
        if row['kdr'] >= 0.7:
            return 1.0
        else:
            return 0.5
    return 0

df['performance'] = df.apply(calculate_performance, axis=1)

print(f"Performance Distribution:")
print(df['performance'].value_counts())
````

---

### Phase 5: Data Filtering

**Objective:** Remove records that don't meet minimum performance thresholds.

**Justification:** Including losses and poor-KDR wins would introduce noise into the recommendation system. Filtering ensures only quality performances influence agent recommendations.


````python
df_filtered = df[df['performance'] > 0].copy()

print(f"Original Records: {len(df)}")
print(f"After Filtering: {len(df_filtered)}")
````

---

### Phase 6: Feature Selection

**Objective:** Select only relevant columns for the prediction model.

**Justification:** Reducing dimensionality to essential features (agent, role, map, performance) improves model efficiency and focuses analysis on business-relevant attributes only.


````python
final_df = df_filtered[['agent', 'role', 'map', 'performance']].copy()

print(f"Final Dataset Shape: {final_df.shape}")
print(final_df)
````

---

### Phase 7: Data Quality Verification

**Objective:** Validate the prepared dataset meets quality standards.

**Justification:** Final verification ensures no missing values, confirms data coverage across agents/maps/roles, and validates the performance score distribution before modeling.


````python
print("Missing Values:")
print(final_df.isnull().sum())

print(f"\nUnique Agents: {final_df['agent'].nunique()}")
print(f"Unique Maps: {final_df['map'].nunique()}")
print(f"Unique Roles: {final_df['role'].nunique()}")

print(f"\nPerformance Score Distribution:")
print(final_df['performance'].value_counts())
````
