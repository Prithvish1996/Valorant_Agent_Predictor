# 3 - Feature importance
In the previous exercise, you trained a Linear Regression model using one or more features to predict car price.

However, some features may not contribute much to the prediction.

In Linear Regression, each feature gets a weight (coefficient).
The size of this weight tells us how important the feature is.

Important:
To compare feature importance fairly, all features must be on the same scale.
If one feature is measured in thousands and another in small numbers, their weights are not directly comparable.

To fix this, we scale the features before training the model again. You can use sklearn’s scale for this(https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.scale.html).
